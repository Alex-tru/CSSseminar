# Who are the Experts?

```python
import pandas as pd
import numpy as np
```

```python
Users=pd.read_csv("../UsersWithCountry.csv")
```

```python
UserTagWeight=pd.read_csv("userFinalTagWeight.csv",sep='\t')
#UserTagWeight
```

```python
UserBadgeWeight=pd.read_csv("../BadgeWeightOfUsers.csv",names=['UserId'])
```

```python
UserBadgeWeight["userId"], UserBadgeWeight["Weight"] = zip(*UserBadgeWeight["UserId"].str.split(',').tolist())
del UserBadgeWeight["UserId"] 
```

```python
UserBadgeWeight=UserBadgeWeight.apply(pd.to_numeric)
```

```python
### DONT RUN BLOCK ###
for i in UserBadgeWeight.index:
    if (i%3000==0):
        print(i,' Done...')
        
    text=str.split(UserBadgeWeight['userId2'].iloc[i],',')
    UserBadgeWeight['userId'].iloc[i]=text[0]
    UserBadgeWeight['badgeWeight'].iloc[i]=text[1]
#=(str.split(str(UserBadgeWeight['userId2']),','))[0]
```

```python
#normalized_df=(df-df.min())/(df.max()-df.min()) feature scaling

UserBadgeWeight["Weight"]=(UserBadgeWeight["Weight"]-UserBadgeWeight["Weight"].min())/(UserBadgeWeight["Weight"].max()-UserBadgeWeight["Weight"].min())
UserTagWeight["tagWeight"]=(UserTagWeight["tagWeight"]-UserTagWeight["tagWeight"].min())/(UserTagWeight["tagWeight"].max()-UserTagWeight["tagWeight"].min())
```

```python
Users["Scaled Reputation"]=0
#Users
```

```python
i=0
for userid in Users['userId'].values:
    
    if i%1000==0:
        print(i,' Done ...')
    
    i=i+1   
    
    Reputation=Users["Reputation"].loc[Users['userId']==userid]
    ScaledBadgeWeight=0 
    scaledTagWeight=0
                                       
    if not (UserBadgeWeight.loc[UserBadgeWeight['userId']==userid].empty):                                    
        ScaledBadgeWeight= UserBadgeWeight["Weight"].loc[UserBadgeWeight['userId']==userid].values[0]
                                       
    if not (UserTagWeight.loc[UserTagWeight['userId']==userid].empty):                                                                      
        ScaledTagWeight= UserTagWeight["tagWeight"].loc[UserTagWeight['userId']==userid].values[0]
                                           
    Users["Scaled Reputation"].loc[Users['userId']==userid]=Reputation*(1+ScaledBadgeWeight)/(1+ScaledTagWeight) 
    
```
