```python
import pandas as pd
```

```python
weights=pd.read_excel('../top3000_tagweights.xlsx')
weights.columns=['Id','TagName','TagWeight']
```

```python
List=weights.to_dict('list')
```

```python
keys=List['TagName']
values=List['TagWeight']
```

```python
dictTagWeight=dict(zip(keys,values))
```

```python
dictTagWeight
```

```python
import pickle

pickle_out=open('dict.pickle','wb')
pickle.dump(dictTagWeight,pickle_out)
pickle_out.close()
```
